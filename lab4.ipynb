# Задание 5. Генератор равномерного распределения
Постройте генератор равномерно распределенного выбора элемента из конечного множества Х, количество элементов которого не известно. Генератор может хранить одновременно только два элемента множества Х. Функция генератора получает множество Х в качестве параметра, например, в виде контейнера. 

```python
import random
import copy
set_count = int(random.uniform(2, 20))
samples = [i for i in range(1, set_count)]
print(samples)
```

```python
generator = []
def generate(samples_):
    samples = copy.copy(samples_)
    item = None
    counter = 2
    generator = samples[:2]
    samples.pop(0)
    samples.pop(0)
    
    while len(samples) > 0:
        counter += 1
        number = int(random.uniform(1, counter))
        if number == generator[0]:
            generator[0] = samples.pop(0)
        else:
            generator[1] = samples.pop(0)
            
    item = generator[0]
    
    return item
```

```python
hist = [generate(samples) for i in range(10000)]
plt.hist(hist, bins=40)
```

Как видно из гистограммы наш метод генерации равномерных-случайных величин равномерен.
